```python
class Calc:
    def __init__(self, a):
        self.a = a
    
    def add(self, b):
        print(self.a + b)
    
    def multiply(self, b):
        print(self.a * b)

calc = Calc(3)
calc.add(4)
calc.multiply(4)
```

```python
class CalcPlus(Calc):
    def subtract(self, b):
        print(self.a - b)
    
    def divide(self, b):
        print(self.a / b)

calc_plus = CalcPlus(3)
calc_plus.add(4)
calc_plus.multiply(4)
calc_plus.subtract(4)
calc_plus.divide(4)
```

```python
x = np.linspace(0, 2*np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.xlabel("x value")
plt.ylabel("y value")

plt.title("sin/cos")

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos", linestyle="dashed")
plt.legend() # 

plt.show()
```
